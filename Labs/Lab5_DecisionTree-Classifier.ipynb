### Example 1

```python
from sklearn import tree
X = [[0, 0], [1, 1], [2,2], [3,3]]
Y = [0, 0, 1, 1]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
clf.predict([[2.5, 2.5]])
```

```python
clf.predict_proba([[2., 2.]])
```

### Example 2

```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
iris.feature_names
```

```python
X[:5]
```

```python
y[:5]
```

```python
# number of classes 
max(y)+1
```

```python
tree.plot_tree(clf)
```

```python
from sklearn.tree import export_text

r = export_text(clf, feature_names=iris['feature_names'])
print(r)
```
