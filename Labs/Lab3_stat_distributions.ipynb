## Question 1 — Gaussian / Normal Distribution

**Task:**
1. Generate a sample of 1000 observations drawn from a Normal distribution with mean `mu = 5` and standard deviation `sigma = 2`.
2. Compute the theoretical Normal probability density function (PDF) over a grid and plot:
   - histogram of the sample (density=True)
   - overlay the theoretical PDF

**In the solution cell you should show code, the plot, and a short interpretation.**

```python
# Solution 1
mu, sigma = 5, 2
sample_norm = np.random.normal(loc=mu, scale=sigma, size=1000)

# Histogram of sample
plt.figure(figsize=(8,4))
plt.hist(sample_norm, bins=30, density=True, alpha=0.6)
# Theoretical PDF
x = np.linspace(sample_norm.min() - 1, sample_norm.max() + 1, 300)
pdf = stats.norm.pdf(x, loc=mu, scale=sigma)
plt.plot(x, pdf, linewidth=2)
plt.title('Histogram of sample with theoretical Normal PDF overlay')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# Short interpretation
mean_emp = sample_norm.mean()
std_emp = sample_norm.std(ddof=1)
print(f'Empirical mean: {mean_emp:.3f}, Empirical std: {std_emp:.3f} (expected mu={mu}, sigma={sigma})')
```
