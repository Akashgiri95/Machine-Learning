### Find the outliers in the data. Update the outliers by mean values of those faetures. 
''' Let's assume that the features are distributed Normally (Gaussian). \\
 Then, the sample values outside +/- 3 *(sigma) of the mean will indicate outliers. \\
 In a zscore, samples with abs(zscore) > 3 are outliers. As such, the threshold could be \\ 
 kept at any value less than 3 also.'''


```python
from scipy import stats
```

```python
# Compute z-score for numeric columns
numeric_cols = data.select_dtypes(include='number').columns
z_scores = data[numeric_cols].apply(stats.zscore)
z_scores
```

```python
threshold = 3

# Replace outliers (|z| > threshold) with column mean
for col in numeric_cols:
    mean_val = data[col].mean()
    outliers_mask = z_scores[col].abs() > threshold
    data.loc[outliers_mask, col] = mean_val

print(data)
```
